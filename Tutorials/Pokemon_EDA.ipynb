```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('../Datasets/Pokemon.csv', index_col=0, encoding='cp1252')
df.head()
```

```python
df.describe()
```

# Scatterplot

```python
# scatterplot
sns.lmplot(x="Attack", y="Defense", data=df)
```

```python
# improved scatterplot
sns.lmplot(x="Attack", y="Defense", data=df, fit_reg=False, hue="Stage")

# matplotlib customization
plt.ylim(0, None)
plt.xlim(0, None)

# color palette
pokemon_palette = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
]
```

# Box Plot

```python
# box plot
stats_df = df.drop(["Total", "Stage", "Legendary"], axis=1)
sns.boxplot(data=stats_df)
```

# Violin Plot

```python
# violin plot
plt.figure(figsize=(10, 5))
sns.set_style("whitegrid")
sns.violinplot(x="Type 1", y="Attack", data=df)
sns.violinplot(x="Type 1", y="Attack", data=df, palette=pokemon_palette)
```

# Swarm Plot

```python
# swarm plot
plt.figure(figsize=(10, 5))
sns.swarmplot(x="Type 1", y="Attack", data=df, palette=pokemon_palette)
```

# Violin-Swarm Plot

```python
# overlayed violin-swarm plot
plt.figure(figsize=(10, 6))
sns.violinplot(x="Type 1", y="Attack", data=df, inner=None, palette=pokemon_palette)
sns.swarmplot(x="Type 1", y="Attack", data=df, color='k', alpha=0.7)
plt.title("Attack by Type")
```

# Heatmap

```python
# create a correlation heatmap of stats
plt.figure(figsize=(10, 5))
corr = stats_df.corr()
sns.heatmap(corr)
```

# Bar Plot

```python
# bar plot
plt.figure(figsize=(10, 5))
sns.countplot(x="Type 1", data=df, palette=pokemon_palette)
plt.xticks(rotation=-45)
```

# Factor Plot

```python
# factor plot
viz = sns.catplot(x="Type 1", y="Attack", data=df, hue="Stage", col="Stage", kind="swarm")
viz.set_xticklabels(rotation=-45)
```

# Density Plot

```python
# density plot
sns.kdeplot(df.Attack, df.Defense)
```

# Distribution Plot

```python
# join distribution plot
sns.jointplot(x="Attack", y="Defense", data=df)
```
